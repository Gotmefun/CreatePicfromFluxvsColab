# Cell 3.0 - üîû Download Models

```python
import os

ChilloutMix = True
RealisticVision = False
DreamShaper = False
CyberRealistic = False

mdir = '/content/drive/MyDrive/SD_Unlimited/models'
os.makedirs(mdir, exist_ok=True)

def dl(fn, url, en, desc):
    if not en:
        return print(f"‚è≠Ô∏è  Skipped: {desc}")
    fp = os.path.join(mdir, fn)
    if os.path.exists(fp):
        size = os.path.getsize(fp) / (1024**3)
        return print(f"‚úÖ Already exists: {desc} ({size:.2f} GB)")
    print(f"\nüì• Downloading: {desc}")
    try:
        !aria2c -x16 -s16 -k1M -c --file-allocation=none --dir="{mdir}" --out="{fn}" "{url}"
        if os.path.exists(fp):
            size = os.path.getsize(fp) / (1024**3)
            print(f"\n‚úÖ Downloaded: {desc} ({size:.2f} GB)\n")
        else:
            print(f"\n‚ùå Failed: {desc}\n")
    except Exception as e:
        print(f"\n‚ùå Error: {e}\n")

print("="*70)
print("üîû Downloading AI Models")
print("="*70 + "\n")

dl("chilloutmix.safetensors", "https://civitai.com/api/download/models/11745", ChilloutMix, "ChilloutMix")
dl("realisticvision.safetensors", "https://civitai.com/api/download/models/130072", RealisticVision, "Realistic Vision")
dl("dreamshaper.safetensors", "https://civitai.com/api/download/models/128713", DreamShaper, "DreamShaper")
dl("cyberrealistic.safetensors", "https://civitai.com/api/download/models/109123", CyberRealistic, "CyberRealistic")

models = [f for f in os.listdir(mdir) if f.endswith('.safetensors')]
print("\n" + "="*70)
print(f"‚úÖ Models ready! ({len(models)} model(s))")
for m in models:
    print(f"   ‚Ä¢ {m}")
print("\nüéØ Next: Cell 4.0")
print("="*70)
```

# Cell 4.0 - üöÄ Start WebUI

## ‚è±Ô∏è This takes 3-5 minutes - BE PATIENT!

**v7 Improvements:**
- ‚è∞ Wait time: 180 seconds (3 minutes)
- üìä Live progress updates
- üîç 60 health check attempts
- ‚úÖ Must see "WebUI is READY!" before Cell 4.5

```python
import os, subprocess, time, threading, urllib.request, json

os.chdir('/content/stable-diffusion-webui')

print("="*70)
print("üöÄ SD Unlimited v7 - Starting WebUI")
print("="*70)
print("\n‚è±Ô∏è  Expected time: 3-5 minutes")
print("‚úÖ CORS: Enabled")
print("‚úÖ API: http://127.0.0.1:7860\n")
print("="*70)

# Start WebUI
process = subprocess.Popen([
    'python', 'launch.py',
    '--skip-python-version-check',
    '--skip-torch-cuda-test',
    '--listen',
    '--port=7860',
    '--api',
    '--api-cors-allow-origins=https://aigen.ptee88.com',
    '--api-cors-allow-origins=*',
    '--enable-insecure-extension-access',
    '--xformers',
    '--no-half-vae',
    '--no-hashing',
    '--lowvram',
    '--skip-install'
], stdout=subprocess.PIPE, stderr=subprocess.STDOUT, text=True, bufsize=1)

print(f"\n‚úÖ Process started (PID: {process.pid})")
print("\nüìã Important logs:\n")
print("-"*70)

# Show logs in background
def show_logs():
    for line in process.stdout:
        line = line.strip()
        if any(kw in line.lower() for kw in ['running', 'loaded', 'model', 'api', 'error', 'warning', 'startup']):
            print(f"  {line}")

log_thread = threading.Thread(target=show_logs, daemon=True)
log_thread.start()

# Wait for WebUI - EXTENDED TO 180 SECONDS
print("\n" + "-"*70)
print("\n‚è≥ Waiting for WebUI (up to 3 minutes)...\n")

max_attempts = 60  # 60 * 3 = 180 seconds
attempt = 0
webui_ready = False

while attempt < max_attempts:
    attempt += 1
    time.sleep(3)
    
    try:
        response = urllib.request.urlopen('http://127.0.0.1:7860', timeout=2)
        try:
            api_response = urllib.request.urlopen('http://127.0.0.1:7860/sdapi/v1/sd-models', timeout=2)
            api_data = json.loads(api_response.read().decode())
            if len(api_data) > 0:
                webui_ready = True
                break
        except:
            pass
        if attempt % 10 == 0:
            elapsed = attempt * 3
            print(f"   ‚è≥ {elapsed}s / 180s - UI up, waiting for API...")
    except:
        if attempt % 10 == 0:
            elapsed = attempt * 3
            print(f"   ‚è≥ {elapsed}s / 180s - Starting up...")

print("\n" + "="*70)

if webui_ready:
    print("\n" + "üéâ"*35)
    print("\n‚úÖ WebUI is READY!")
    print("‚úÖ API: http://127.0.0.1:7860")
    print("‚úÖ CORS: Enabled\n")
    print("üéâ"*35)
    
    try:
        api_response = urllib.request.urlopen('http://127.0.0.1:7860/sdapi/v1/sd-models', timeout=5)
        models = json.loads(api_response.read().decode())
        print(f"\nüì¶ Loaded Models: {len(models)}")
        for m in models:
            print(f"   ‚Ä¢ {m.get('title', m.get('model_name', 'Unknown'))}")
    except Exception as e:
        print(f"\n‚ö†Ô∏è  API test failed: {e}")
    
    print("\n" + "="*70)
    print("üéØ Next: Cell 4.5 (Debug Helper)")
    print("‚ö†Ô∏è  DO NOT skip Cell 4.5!")
    print("="*70)
else:
    print("\n‚ùå WebUI failed to start in 3 minutes")
    print("\nüîß Solutions:")
    print("   1. Check logs above for errors")
    print("   2. Runtime > Restart runtime")
    print("   3. Make sure GPU is T4")
    print("   4. Try running Cell 4.0 again")
    print("\n" + "="*70)
```
