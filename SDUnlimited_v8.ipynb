# Cell 3.0 - üîû Download Models

```python
import os

# Model selection - ‡πÄ‡∏õ‡∏•‡∏µ‡πà‡∏¢‡∏ô‡πÄ‡∏õ‡πá‡∏ô True ‡πÄ‡∏û‡∏∑‡πà‡∏≠‡∏î‡∏≤‡∏ß‡∏ô‡πå‡πÇ‡∏´‡∏•‡∏î
ChilloutMix = True
RealisticVision = False
DreamShaper = False
CyberRealistic = False

mdir = '/content/drive/MyDrive/SD_Unlimited/models'
os.makedirs(mdir, exist_ok=True)

def dl(fn, url, en, desc):
    if not en:
        return print(f"‚è≠Ô∏è  Skipped: {desc}")
    
    fp = os.path.join(mdir, fn)
    if os.path.exists(fp):
        size = os.path.getsize(fp) / (1024**3)
        return print(f"‚úÖ Already exists: {desc} ({size:.2f} GB)")
    
    print(f"\n{'='*70}")
    print(f"üì• Downloading: {desc}")
    print(f"{'='*70}")
    
    try:
        !aria2c -x16 -s16 -k1M -c --file-allocation=none --dir="{mdir}" --out="{fn}" "{url}"
        
        if os.path.exists(fp):
            size = os.path.getsize(fp) / (1024**3)
            print(f"\n‚úÖ Downloaded: {desc} ({size:.2f} GB)")
        else:
            print(f"\n‚ùå Failed: {desc} - File not found after download")
    except Exception as e:
        print(f"\n‚ùå Error downloading {desc}: {e}")

print("="*70)
print("üîû Downloading AI Models")
print("="*70 + "\n")

dl("chilloutmix.safetensors", 
   "https://civitai.com/api/download/models/11745", 
   ChilloutMix, 
   "ChilloutMix (Realistic)")

dl("realisticvision.safetensors", 
   "https://civitai.com/api/download/models/130072", 
   RealisticVision, 
   "Realistic Vision v5.1")

dl("dreamshaper.safetensors", 
   "https://civitai.com/api/download/models/128713", 
   DreamShaper, 
   "DreamShaper v8")

dl("cyberrealistic.safetensors", 
   "https://civitai.com/api/download/models/109123", 
   CyberRealistic, 
   "CyberRealistic v4.1")

# Summary
models = [f for f in os.listdir(mdir) if f.endswith('.safetensors')]
print("\n" + "="*70)
print(f"‚úÖ Models ready! ({len(models)} model(s))")
print("="*70)

if len(models) == 0:
    print("\n‚ö†Ô∏è  WARNING: No models found!")
    print("   Please enable at least one model above and rerun this cell")
else:
    for m in models:
        size = os.path.getsize(os.path.join(mdir, m)) / (1024**3)
        print(f"   ‚Ä¢ {m} ({size:.2f} GB)")
    print("\nüéØ Next: Cell 4.0")

print("="*70)
```

## üÜï v8 Features:

- ‚úÖ Real-time log streaming (‡πÑ‡∏°‡πà‡∏ï‡πâ‡∏≠‡∏á‡∏£‡∏≠)

```python
import os
import subprocess
import time
import threading
import urllib.request
import json
import base64
from queue import Queue, Empty

os.chdir('/content/stable-diffusion-webui')

print("="*70)
print("üöÄ SD Unlimited v8 - Starting WebUI")
print("="*70)
print("\n‚è±Ô∏è  Expected time: 3-5 minutes")
print("‚úÖ CORS: Enabled (wildcard)")
print("‚úÖ API: http://127.0.0.1:7860")
print("‚úÖ Real-time logs: Enabled\n")
print("="*70)
print("\nüìã Starting WebUI process...\n")

# Start WebUI with proper configuration
process = subprocess.Popen([
    'python', 'launch.py',
    '--skip-python-version-check',
    '--skip-torch-cuda-test',
    '--skip-install',
    '--listen',
    '--port=7860',
    '--api',
    '--api-cors-allow-origins=*',
    '--enable-insecure-extension-access',
    '--xformers',
    '--no-half-vae',
    '--no-hashing',
    '--lowvram'
], 
stdout=subprocess.PIPE, 
stderr=subprocess.STDOUT, 
text=True, 
bufsize=1,
universal_newlines=True)

print(f"‚úÖ Process started (PID: {process.pid})")
print("\n" + "-"*70)
print("üìã Live Logs:")
print("-"*70 + "\n")

# Track important milestones
milestones = {
    'startup': False,
    'loading_model': False,
    'model_loaded': False,
    'api_ready': False,
    'ui_ready': False
}

# Real-time log processing
def process_logs(proc, milestones):
    """Process logs in real-time and track milestones"""
    for line in iter(proc.stdout.readline, ''):
        if not line:
            break
        
        line = line.strip()
        if not line:
            continue
        
        # Track milestones
        line_lower = line.lower()
        
        if 'loading model' in line_lower or 'loading checkpoint' in line_lower:
            milestones['loading_model'] = True
            print(f"\nüîÑ {line}")
        
        elif 'model loaded' in line_lower or 'checkpoint loaded' in line_lower:
            milestones['model_loaded'] = True
            print(f"\n‚úÖ {line}")
        
        elif 'running on local url' in line_lower or 'uvicorn running' in line_lower:
            milestones['ui_ready'] = True
            print(f"\n‚úÖ {line}")
        
        elif 'startup time' in line_lower:
            print(f"\n‚è±Ô∏è  {line}")
        
        elif any(word in line_lower for word in ['error', 'failed', 'exception']):
            print(f"\n‚ùå ERROR: {line}")
        
        elif any(word in line_lower for word in ['warning']):
            print(f"\n‚ö†Ô∏è  {line}")
        
        elif any(word in line_lower for word in ['installing', 'downloading', 'applying']):
            print(f"   {line}")

# Start log processing in background
log_thread = threading.Thread(target=process_logs, args=(process, milestones), daemon=True)
log_thread.start()

# Smart health check with multiple stages
print("\n" + "="*70)
print("üîç Waiting for WebUI (Smart Health Check)")
print("="*70 + "\n")

max_wait = 300  # 5 minutes max
start_time = time.time()
stage = 1

while time.time() - start_time < max_wait:
    elapsed = int(time.time() - start_time)
    
    # Check if process died
    if process.poll() is not None:
        print(f"\n‚ùå ERROR: WebUI process died!")
        print(f"   Exit code: {process.returncode}")
        print("\nüîß Solutions:")
        print("   1. Check logs above for errors")
        print("   2. Runtime ‚Üí Restart runtime")
        print("   3. Rerun Cell 1.0 and 3.0")
        raise Exception("Process died")
    
    # Stage 1: Wait for port to open (60s max)
    if stage == 1:
        try:
            urllib.request.urlopen('http://127.0.0.1:7860', timeout=2)
            print(f"\n‚úÖ Stage 1/4: Port 7860 is open ({elapsed}s)")
            stage = 2
        except:
            if elapsed % 10 == 0:
                print(f"   ‚è≥ Stage 1/4: Waiting for port... ({elapsed}s)")
            time.sleep(2)
    
    # Stage 2: Wait for API endpoint (30s max)
    elif stage == 2:
        try:
            response = urllib.request.urlopen('http://127.0.0.1:7860/sdapi/v1/sd-models', timeout=2)
            print(f"‚úÖ Stage 2/4: API endpoint responding ({elapsed}s)")
            stage = 3
        except:
            if elapsed % 10 == 0:
                print(f"   ‚è≥ Stage 2/4: Waiting for API... ({elapsed}s)")
            time.sleep(2)
    
    # Stage 3: Check if models loaded (60s max)
    elif stage == 3:
        try:
            response = urllib.request.urlopen('http://127.0.0.1:7860/sdapi/v1/sd-models', timeout=3)
            models = json.loads(response.read().decode())
            
            if len(models) > 0:
                print(f"‚úÖ Stage 3/4: Model loaded ({len(models)} model(s)) ({elapsed}s)")
                for m in models:
                    print(f"   ‚Ä¢ {m.get('title', m.get('model_name', 'Unknown'))}")
                stage = 4
            else:
                if elapsed % 10 == 0:
                    print(f"   ‚è≥ Stage 3/4: Waiting for model to load... ({elapsed}s)")
                time.sleep(3)
        except:
            time.sleep(2)
    
    # Stage 4: Test generation
    elif stage == 4:
        print(f"\nüß™ Stage 4/4: Testing image generation...")
        try:
            test_payload = {
                "prompt": "a red apple",
                "steps": 1,
                "width": 128,
                "height": 128
            }
            
            req = urllib.request.Request(
                'http://127.0.0.1:7860/sdapi/v1/txt2img',
                data=json.dumps(test_payload).encode('utf-8'),
                headers={'Content-Type': 'application/json'}
            )
            
            response = urllib.request.urlopen(req, timeout=30)
            result = json.loads(response.read().decode())
            
            if 'images' in result and len(result['images']) > 0:
                print(f"‚úÖ Stage 4/4: Test generation successful! ({elapsed}s)")
                break
            else:
                print(f"‚ö†Ô∏è  Stage 4/4: No images in response")
                time.sleep(5)
                
        except Exception as e:
            print(f"   ‚è≥ Stage 4/4: Generation test failed: {e}")
            print(f"   Retrying... ({elapsed}s)")
            time.sleep(5)

# Final verification
print("\n" + "="*70)

if stage == 4:
    try:
        # Verify one more time
        response = urllib.request.urlopen('http://127.0.0.1:7860/sdapi/v1/sd-models', timeout=5)
        models = json.loads(response.read().decode())
        
        print("\n" + "üéâ"*35)
        print("\n‚úÖ‚úÖ‚úÖ WebUI is READY! ‚úÖ‚úÖ‚úÖ")
        print("\nüéâ"*35)
        print(f"\n‚úÖ API: http://127.0.0.1:7860")
        print(f"‚úÖ CORS: Enabled (*)")
        print(f"‚úÖ Models: {len(models)}")
        for m in models:
            print(f"   ‚Ä¢ {m.get('title', m.get('model_name', 'Unknown'))}")
        print(f"‚úÖ Test Generation: Passed")
        print(f"‚úÖ Total startup time: {elapsed}s")
        
        print("\n" + "="*70)
        print("üéØ Next: Run Cell 5.0 to create public URL")
        print("="*70)
        
    except Exception as e:
        print(f"\n‚ö†Ô∏è  Final verification failed: {e}")
        print("\nWebUI might be unstable. Try running Cell 5.0 anyway.")
else:
    print(f"\n‚ùå WebUI failed to start completely")
    print(f"   Reached stage {stage}/4 in {elapsed}s")
    print("\nüîß Solutions:")
    print("   1. Check logs above for specific errors")
    print("   2. Runtime ‚Üí Restart runtime")
    print("   3. Rerun all cells from Cell 1.0")
    print("\n" + "="*70)
```
