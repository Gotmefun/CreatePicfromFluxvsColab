# Cell 3.0 - ğŸ” Download Models

```python
import os

# Model selection - à¹€à¸›à¸¥à¸µà¹ˆà¸¢à¸™à¹€à¸›à¹‡à¸™ True à¹€à¸à¸·à¹ˆà¸­à¸”à¸²à¸§à¸™à¹Œà¹‚à¸«à¸¥à¸”
ChilloutMix = True
RealisticVision = False
DreamShaper = False
CyberRealistic = False

mdir = '/content/drive/MyDrive/SD_Unlimited/models'
os.makedirs(mdir, exist_ok=True)

def dl(fn, url, en, desc):
    if not en:
        return print(f"â­ï¸  Skipped: {desc}")
    
    fp = os.path.join(mdir, fn)
    if os.path.exists(fp):
        size = os.path.getsize(fp) / (1024**3)
        return print(f"âœ… Already exists: {desc} ({size:.2f} GB)")
    
    print(f"\n{'='*70}")
    print(f"ğŸ“¥ Downloading: {desc}")
    print(f"{'='*70}")
    
    try:
        !aria2c -x16 -s16 -k1M -c --file-allocation=none --dir="{mdir}" --out="{fn}" "{url}"
        
        if os.path.exists(fp):
            size = os.path.getsize(fp) / (1024**3)
            print(f"\nâœ… Downloaded: {desc} ({size:.2f} GB)")
        else:
            print(f"\nâŒ Failed: {desc} - File not found after download")
    except Exception as e:
        print(f"\nâŒ Error downloading {desc}: {e}")

print("="*70)
print("ğŸ” Downloading AI Models")
print("="*70 + "\n")

dl("chilloutmix.safetensors", 
   "https://civitai.com/api/download/models/11745", 
   ChilloutMix, 
   "ChilloutMix (Realistic)")

dl("realisticvision.safetensors", 
   "https://civitai.com/api/download/models/130072", 
   RealisticVision, 
   "Realistic Vision v5.1")

dl("dreamshaper.safetensors", 
   "https://civitai.com/api/download/models/128713", 
   DreamShaper, 
   "DreamShaper v8")

dl("cyberrealistic.safetensors", 
   "https://civitai.com/api/download/models/109123", 
   CyberRealistic, 
   "CyberRealistic v4.1")

# Summary
models = [f for f in os.listdir(mdir) if f.endswith('.safetensors')]
print("\n" + "="*70)
print(f"âœ… Models ready! ({len(models)} model(s))")
print("="*70)

if len(models) == 0:
    print("\nâš ï¸  WARNING: No models found!")
    print("   Please enable at least one model above and rerun this cell")
else:
    for m in models:
        size = os.path.getsize(os.path.join(mdir, m)) / (1024**3)
        print(f"   â€¢ {m} ({size:.2f} GB)")
    print("\nğŸ¯ Next: Cell 4.0")

print("="*70)
```

## ğŸ†• v8 Features:

```python
import os
import time
import urllib.request
import json

# Check if WebUI folder exists
if not os.path.exists('/content/stable-diffusion-webui'):
    print("="*70)
    print("âŒ ERROR: WebUI folder not found!")
    print("="*70)
    print("\nğŸ”§ Solution: Run Cell 1.0 first!")
    print("="*70)
    raise FileNotFoundError("WebUI not found. Please run Cell 1.0 first.")

os.chdir('/content/stable-diffusion-webui')

print("="*70)
print("ğŸš€ SD Unlimited v9 - Starting WebUI")
print("="*70)
print("\nâ±ï¸  Expected time: 5-10 minutes (first run)")
print("âœ… CORS: Enabled")
print("âœ… API: http://127.0.0.1:7860\n")
print("="*70)

# Start WebUI in background using nohup
print("\nğŸ“‹ Starting WebUI in background...\n")
print("="*70)
print("ğŸ“º WebUI Logs (live):")
print("="*70 + "\n")

# Use nohup to run in background and tee to show output
!nohup python launch.py \
  --skip-python-version-check \
  --skip-torch-cuda-test \
  --listen \
  --port=7860 \
  --api \
  --api-cors-allow-origins=* \
  --enable-insecure-extension-access \
  --xformers \
  --no-half-vae \
  --no-hashing \
  --lowvram \
  > /tmp/webui.log 2>&1 &

import subprocess
import signal

print("\n" + "="*70)
print("âœ… WebUI process started in background")
print("="*70)

# Give it a moment to start
time.sleep(5)

# Show logs
print("\nğŸ“‹ Showing live logs...\n")
print("="*70)

# Tail logs in background
log_proc = subprocess.Popen(
    ['tail', '-f', '/tmp/webui.log'],
    stdout=subprocess.PIPE,
    stderr=subprocess.PIPE,
    text=True
)

# Health check
print("\n" + "="*70)
print("ğŸ” Health Check (waiting for WebUI...)")
print("="*70 + "\n")

max_wait = 600  # 10 minutes
start_time = time.time()
stage = 1
last_log_check = 0

while time.time() - start_time < max_wait:
    elapsed = int(time.time() - start_time)
    
    # Show log snippets every 30 seconds
    if elapsed - last_log_check >= 30:
        try:
            with open('/tmp/webui.log', 'r') as f:
                lines = f.readlines()
                if lines:
                    recent = lines[-5:]  # Last 5 lines
                    print(f"\nğŸ“ Recent logs ({elapsed}s):")
                    for line in recent:
                        line = line.strip()
                        if line:
                            print(f"   {line}")
        except:
            pass
        last_log_check = elapsed
    
    # Stage 1: Wait for port
    if stage == 1:
        try:
            urllib.request.urlopen('http://127.0.0.1:7860', timeout=2)
            print(f"\nâœ… Stage 1/4: Port 7860 is OPEN ({elapsed}s)")
            stage = 2
        except:
            if elapsed % 30 == 0:
                print(f"   â³ Stage 1/4: Waiting for port 7860... ({elapsed}s / {max_wait}s)")
            time.sleep(3)
    
    # Stage 2: API endpoint
    elif stage == 2:
        try:
            response = urllib.request.urlopen('http://127.0.0.1:7860/sdapi/v1/sd-models', timeout=3)
            print(f"âœ… Stage 2/4: API endpoint responding ({elapsed}s)")
            stage = 3
        except:
            if elapsed % 10 == 0:
                print(f"   â³ Stage 2/4: Waiting for API... ({elapsed}s)")
            time.sleep(2)
    
    # Stage 3: Models loaded
    elif stage == 3:
        try:
            response = urllib.request.urlopen('http://127.0.0.1:7860/sdapi/v1/sd-models', timeout=3)
            models = json.loads(response.read().decode())
            
            if len(models) > 0:
                print(f"âœ… Stage 3/4: Model loaded ({len(models)} model(s)) ({elapsed}s)")
                for m in models:
                    print(f"   â€¢ {m.get('title', m.get('model_name', 'Unknown'))}")
                stage = 4
            else:
                if elapsed % 10 == 0:
                    print(f"   â³ Stage 3/4: Waiting for model... ({elapsed}s)")
                time.sleep(3)
        except:
            time.sleep(2)
    
    # Stage 4: Test generation
    elif stage == 4:
        print(f"\nğŸ§ª Stage 4/4: Testing image generation...")
        try:
            test_payload = {
                "prompt": "test",
                "steps": 1,
                "width": 128,
                "height": 128
            }
            
            req = urllib.request.Request(
                'http://127.0.0.1:7860/sdapi/v1/txt2img',
                data=json.dumps(test_payload).encode('utf-8'),
                headers={'Content-Type': 'application/json'}
            )
            
            response = urllib.request.urlopen(req, timeout=30)
            result = json.loads(response.read().decode())
            
            if 'images' in result and len(result['images']) > 0:
                print(f"âœ… Stage 4/4: Generation successful! ({elapsed}s)")
                break
            else:
                print(f"âš ï¸  No images, retrying...")
                time.sleep(5)
                
        except Exception as e:
            print(f"   â³ Generation test failed: {str(e)[:50]}")
            time.sleep(5)

# Stop log tail
log_proc.terminate()

# Final result
print("\n" + "="*70)

if stage >= 4:
    try:
        response = urllib.request.urlopen('http://127.0.0.1:7860/sdapi/v1/sd-models', timeout=5)
        models = json.loads(response.read().decode())
        
        print("\n" + "ğŸ‰"*35)
        print("\nâœ…âœ…âœ… WebUI is READY! âœ…âœ…âœ…")
        print("\nğŸ‰"*35)
        print(f"\nâœ… API: http://127.0.0.1:7860")
        print(f"âœ… Models: {len(models)}")
        for m in models:
            print(f"   â€¢ {m.get('title', m.get('model_name', 'Unknown'))}")
        print(f"âœ… Total time: {elapsed}s")
        
        print("\n" + "="*70)
        print("ğŸ¯ Next: Run Cell 5.0")
        print("="*70)
        
    except Exception as e:
        print(f"âš ï¸  Verification failed: {e}")
else:
    print(f"\nâŒ WebUI failed to start")
    print(f"   Stuck at stage {stage}/4 after {elapsed}s")
    print("\nğŸ“‹ Check full logs:")
    print("   Run: !tail -100 /tmp/webui.log")
    print("\nğŸ”§ Common issues:")
    print("   â€¢ Out of GPU memory â†’ Use T4 GPU")
    print("   â€¢ No models â†’ Rerun Cell 3.0")
    print("   â€¢ Dependency error â†’ Runtime â†’ Restart runtime")
    print("\n" + "="*70)
```
