# Cell 3.0 - üîû Download Models

```python
import os

# Model selection - ‡πÄ‡∏õ‡∏•‡∏µ‡πà‡∏¢‡∏ô‡πÄ‡∏õ‡πá‡∏ô True ‡πÄ‡∏û‡∏∑‡πà‡∏≠‡∏î‡∏≤‡∏ß‡∏ô‡πå‡πÇ‡∏´‡∏•‡∏î
ChilloutMix = True
RealisticVision = False
DreamShaper = False
CyberRealistic = False

mdir = '/content/drive/MyDrive/SD_Unlimited/models'
os.makedirs(mdir, exist_ok=True)

def dl(fn, url, en, desc):
    if not en:
        return print(f"‚è≠Ô∏è  Skipped: {desc}")
    
    fp = os.path.join(mdir, fn)
    if os.path.exists(fp):
        size = os.path.getsize(fp) / (1024**3)
        return print(f"‚úÖ Already exists: {desc} ({size:.2f} GB)")
    
    print(f"\n{'='*70}")
    print(f"üì• Downloading: {desc}")
    print(f"{'='*70}")
    
    try:
        !aria2c -x16 -s16 -k1M -c --file-allocation=none --dir="{mdir}" --out="{fn}" "{url}"
        
        if os.path.exists(fp):
            size = os.path.getsize(fp) / (1024**3)
            print(f"\n‚úÖ Downloaded: {desc} ({size:.2f} GB)")
        else:
            print(f"\n‚ùå Failed: {desc} - File not found after download")
    except Exception as e:
        print(f"\n‚ùå Error downloading {desc}: {e}")

print("="*70)
print("üîû Downloading AI Models")
print("="*70 + "\n")

dl("chilloutmix.safetensors", 
   "https://civitai.com/api/download/models/11745", 
   ChilloutMix, 
   "ChilloutMix (Realistic)")

dl("realisticvision.safetensors", 
   "https://civitai.com/api/download/models/130072", 
   RealisticVision, 
   "Realistic Vision v5.1")

dl("dreamshaper.safetensors", 
   "https://civitai.com/api/download/models/128713", 
   DreamShaper, 
   "DreamShaper v8")

dl("cyberrealistic.safetensors", 
   "https://civitai.com/api/download/models/109123", 
   CyberRealistic, 
   "CyberRealistic v4.1")

# Summary
models = [f for f in os.listdir(mdir) if f.endswith('.safetensors')]
print("\n" + "="*70)
print(f"‚úÖ Models ready! ({len(models)} model(s))")
print("="*70)

if len(models) == 0:
    print("\n‚ö†Ô∏è  WARNING: No models found!")
    print("   Please enable at least one model above and rerun this cell")
else:
    for m in models:
        size = os.path.getsize(os.path.join(mdir, m)) / (1024**3)
        print(f"   ‚Ä¢ {m} ({size:.2f} GB)")
    print("\nüéØ Next: Cell 4.0")

print("="*70)
```

# Cell 4.0 - üöÄ Start WebUI with Real-time Monitoring

## üÜï v8 Features:

- ‚úÖ Real-time log streaming (‡πÑ‡∏°‡πà‡∏ï‡πâ‡∏≠‡∏á‡∏£‡∏≠)

```python
import os
import time
import urllib.request
import json

# Check if WebUI folder exists
if not os.path.exists('/content/stable-diffusion-webui'):
    print("="*70)
    print("‚ùå ERROR: WebUI folder not found!")
    print("="*70)
    print("\nüîß Solution: Run Cell 1.0 first!")
    print("="*70)
    raise FileNotFoundError("WebUI not found. Please run Cell 1.0 first.")

os.chdir('/content/stable-diffusion-webui')

print("="*70)
print("üöÄ SD Unlimited v9 - Starting WebUI")
print("="*70)
print("\n‚è±Ô∏è  Expected time: 5-10 minutes (first run)")
print("‚úÖ CORS: Enabled")
print("‚úÖ API: http://127.0.0.1:7860\n")
print("="*70)

# Start WebUI in background using nohup
print("\nüìã Starting WebUI in background...\n")
print("="*70)
print("üì∫ WebUI Logs (live):")
print("="*70 + "\n")

# Use nohup to run in background and tee to show output
!nohup python launch.py \
  --skip-python-version-check \
  --skip-torch-cuda-test \
  --listen \
  --port=7860 \
  --api \
  --api-cors-allow-origins=* \
  --enable-insecure-extension-access \
  --xformers \
  --no-half-vae \
  --no-hashing \
  --lowvram \
  > /tmp/webui.log 2>&1 &

import subprocess
import signal

print("\n" + "="*70)
print("‚úÖ WebUI process started in background")
print("="*70)

# Give it a moment to start
time.sleep(5)

# Show logs
print("\nüìã Showing live logs...\n")
print("="*70)

# Tail logs in background
log_proc = subprocess.Popen(
    ['tail', '-f', '/tmp/webui.log'],
    stdout=subprocess.PIPE,
    stderr=subprocess.PIPE,
    text=True
)

# Health check
print("\n" + "="*70)
print("üîç Health Check (waiting for WebUI...)")
print("="*70 + "\n")

max_wait = 600  # 10 minutes
start_time = time.time()
stage = 1
last_log_check = 0

while time.time() - start_time < max_wait:
    elapsed = int(time.time() - start_time)
    
    # Show log snippets every 30 seconds
    if elapsed - last_log_check >= 30:
        try:
            with open('/tmp/webui.log', 'r') as f:
                lines = f.readlines()
                if lines:
                    recent = lines[-5:]  # Last 5 lines
                    print(f"\nüìù Recent logs ({elapsed}s):")
                    for line in recent:
                        line = line.strip()
                        if line:
                            print(f"   {line}")
        except:
            pass
        last_log_check = elapsed
    
    # Stage 1: Wait for port
    if stage == 1:
        try:
            urllib.request.urlopen('http://127.0.0.1:7860', timeout=2)
            print(f"\n‚úÖ Stage 1/4: Port 7860 is OPEN ({elapsed}s)")
            stage = 2
        except:
            if elapsed % 30 == 0:
                print(f"   ‚è≥ Stage 1/4: Waiting for port 7860... ({elapsed}s / {max_wait}s)")
            time.sleep(3)
    
    # Stage 2: API endpoint
    elif stage == 2:
        try:
            response = urllib.request.urlopen('http://127.0.0.1:7860/sdapi/v1/sd-models', timeout=3)
            print(f"‚úÖ Stage 2/4: API endpoint responding ({elapsed}s)")
            stage = 3
        except:
            if elapsed % 10 == 0:
                print(f"   ‚è≥ Stage 2/4: Waiting for API... ({elapsed}s)")
            time.sleep(2)
    
    # Stage 3: Models loaded
    elif stage == 3:
        try:
            response = urllib.request.urlopen('http://127.0.0.1:7860/sdapi/v1/sd-models', timeout=3)
            models = json.loads(response.read().decode())
            
            if len(models) > 0:
                print(f"‚úÖ Stage 3/4: Model loaded ({len(models)} model(s)) ({elapsed}s)")
                for m in models:
                    print(f"   ‚Ä¢ {m.get('title', m.get('model_name', 'Unknown'))}")
                stage = 4
            else:
                if elapsed % 10 == 0:
                    print(f"   ‚è≥ Stage 3/4: Waiting for model... ({elapsed}s)")
                time.sleep(3)
        except:
            time.sleep(2)
    
    # Stage 4: Test generation
    elif stage == 4:
        print(f"\nüß™ Stage 4/4: Testing image generation...")
        try:
            test_payload = {
                "prompt": "test",
                "steps": 1,
                "width": 128,
                "height": 128
            }
            
            req = urllib.request.Request(
                'http://127.0.0.1:7860/sdapi/v1/txt2img',
                data=json.dumps(test_payload).encode('utf-8'),
                headers={'Content-Type': 'application/json'}
            )
            
            response = urllib.request.urlopen(req, timeout=30)
            result = json.loads(response.read().decode())
            
            if 'images' in result and len(result['images']) > 0:
                print(f"‚úÖ Stage 4/4: Generation successful! ({elapsed}s)")
                break
            else:
                print(f"‚ö†Ô∏è  No images, retrying...")
                time.sleep(5)
                
        except Exception as e:
            print(f"   ‚è≥ Generation test failed: {str(e)[:50]}")
            time.sleep(5)

# Stop log tail
log_proc.terminate()

# Final result
print("\n" + "="*70)

if stage >= 4:
    try:
        response = urllib.request.urlopen('http://127.0.0.1:7860/sdapi/v1/sd-models', timeout=5)
        models = json.loads(response.read().decode())
        
        print("\n" + "üéâ"*35)
        print("\n‚úÖ‚úÖ‚úÖ WebUI is READY! ‚úÖ‚úÖ‚úÖ")
        print("\nüéâ"*35)
        print(f"\n‚úÖ API: http://127.0.0.1:7860")
        print(f"‚úÖ Models: {len(models)}")
        for m in models:
            print(f"   ‚Ä¢ {m.get('title', m.get('model_name', 'Unknown'))}")
        print(f"‚úÖ Total time: {elapsed}s")
        
        print("\n" + "="*70)
        print("üéØ Next: Run Cell 5.0")
        print("="*70)
        
    except Exception as e:
        print(f"‚ö†Ô∏è  Verification failed: {e}")
else:
    print(f"\n‚ùå WebUI failed to start")
    print(f"   Stuck at stage {stage}/4 after {elapsed}s")
    print("\nüìã Check full logs:")
    print("   Run: !tail -100 /tmp/webui.log")
    print("\nüîß Common issues:")
    print("   ‚Ä¢ Out of GPU memory ‚Üí Use T4 GPU")
    print("   ‚Ä¢ No models ‚Üí Rerun Cell 3.0")
    print("   ‚Ä¢ Dependency error ‚Üí Runtime ‚Üí Restart runtime")
    print("\n" + "="*70)
```
