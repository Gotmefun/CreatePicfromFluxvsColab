```python
#@title ## üõ°Ô∏è Cell 1: Anti-Timeout Setup (‡∏£‡∏±‡∏ô‡∏Ñ‡∏£‡∏±‡πâ‡∏á‡πÄ‡∏î‡∏µ‡∏¢‡∏ß‡∏ï‡∏≠‡∏ô‡πÄ‡∏£‡∏¥‡πà‡∏°‡∏ï‡πâ‡∏ô) { display-mode: "form" }

#@markdown ---
#@markdown ### ‡∏õ‡πâ‡∏≠‡∏á‡∏Å‡∏±‡∏ô Colab Timeout ‡∏≠‡∏±‡∏ï‡πÇ‡∏ô‡∏°‡∏±‡∏ï‡∏¥
#@markdown Script ‡∏ô‡∏µ‡πâ‡∏à‡∏∞‡∏Ñ‡∏•‡∏¥‡∏Å‡∏õ‡∏∏‡πà‡∏° "Connect" ‡∏ó‡∏∏‡∏Å 5 ‡∏ô‡∏≤‡∏ó‡∏µ‡πÄ‡∏û‡∏∑‡πà‡∏≠‡πÑ‡∏°‡πà‡πÉ‡∏´‡πâ Colab ‡∏Ñ‡∏¥‡∏î‡∏ß‡πà‡∏≤‡∏Ñ‡∏∏‡∏ì‡πÑ‡∏°‡πà‡πÑ‡∏î‡πâ‡πÉ‡∏ä‡πâ‡∏á‡∏≤‡∏ô

from IPython.display import display, HTML, Javascript
import time

# Anti-Timeout JavaScript
anti_timeout_script = """
<script>
// Anti-Timeout Protection
console.log('%cüõ°Ô∏è Anti-Timeout Protection Activated!', 'color: green; font-size: 20px; font-weight: bold;');

var antiTimeoutInterval = null;
var keepAliveCount = 0;
var startTime = new Date();

function clickConnect() {
  try {
    // ‡∏Ñ‡∏•‡∏¥‡∏Å‡∏õ‡∏∏‡πà‡∏° Connect ‡∏ñ‡πâ‡∏≤‡∏°‡∏µ
    var connectButton = document.querySelector('colab-connect-button');
    if (connectButton) {
      connectButton.shadowRoot.querySelector('#connect').click();
      console.log('üîå Auto-clicked Connect button');
    }
    
    // ‡∏Ñ‡∏•‡∏¥‡∏Å‡∏ó‡∏µ‡πà output area ‡πÄ‡∏û‡∏∑‡πà‡∏≠‡∏™‡∏£‡πâ‡∏≤‡∏á interaction
    var outputs = document.querySelectorAll('.output');
    if (outputs.length > 0) {
      outputs[0].click();
    }
    
    keepAliveCount++;
    var uptime = Math.floor((new Date() - startTime) / 60000); // ‡∏ô‡∏≤‡∏ó‡∏µ
    console.log('‚úÖ Keep-Alive #' + keepAliveCount + ' | Uptime: ' + uptime + ' minutes');
    
    // ‡∏≠‡∏±‡∏û‡πÄ‡∏î‡∏ï‡∏™‡∏ñ‡∏≤‡∏ô‡∏∞‡∏ö‡∏ô UI
    updateStatus(uptime, keepAliveCount);
    
  } catch(e) {
    console.log('‚ö†Ô∏è Keep-alive error:', e);
  }
}

function updateStatus(uptime, count) {
  var statusDiv = document.getElementById('anti-timeout-status');
  if (statusDiv) {
    statusDiv.innerHTML = `
      <div style="background: linear-gradient(135deg, #667eea 0%, #764ba2 100%); padding: 20px; border-radius: 10px; color: white; margin: 10px 0; font-family: monospace;">
        <h3 style="margin: 0 0 10px 0;">üõ°Ô∏è Anti-Timeout Status</h3>
        <div style="display: grid; grid-template-columns: 1fr 1fr; gap: 10px;">
          <div style="background: rgba(255,255,255,0.2); padding: 10px; border-radius: 5px;">
            <div style="font-size: 12px; opacity: 0.8;">Uptime</div>
            <div style="font-size: 24px; font-weight: bold;">${uptime} min</div>
          </div>
          <div style="background: rgba(255,255,255,0.2); padding: 10px; border-radius: 5px;">
            <div style="font-size: 12px; opacity: 0.8;">Keep-Alive</div>
            <div style="font-size: 24px; font-weight: bold;">#${count}</div>
          </div>
        </div>
        <div style="margin-top: 10px; font-size: 12px; opacity: 0.9;">
          ‚è∞ Next ping in 5 minutes | üü¢ Status: Active
        </div>
      </div>
    `;
  }
}

// ‡πÄ‡∏£‡∏¥‡πà‡∏°‡∏ï‡πâ‡∏ô Anti-Timeout
if (antiTimeoutInterval) {
  clearInterval(antiTimeoutInterval);
}

// ‡∏Ñ‡∏•‡∏¥‡∏Å‡∏ó‡∏±‡∏ô‡∏ó‡∏µ‡πÄ‡∏°‡∏∑‡πà‡∏≠‡πÄ‡∏£‡∏¥‡πà‡∏°‡∏ï‡πâ‡∏ô
clickConnect();

// ‡∏ï‡∏±‡πâ‡∏á interval ‡∏ó‡∏∏‡∏Å 5 ‡∏ô‡∏≤‡∏ó‡∏µ (300000 ms)
antiTimeoutInterval = setInterval(clickConnect, 300000);

console.log('‚è∞ Keep-alive interval set to 5 minutes');
console.log('üí° Tip: Keep this tab open (can minimize browser)');
</script>

<div id="anti-timeout-status"></div>

<div style="background: #10b981; padding: 15px; border-radius: 8px; color: white; margin: 10px 0;">
  <h3 style="margin: 0;">‚úÖ Anti-Timeout Protection Enabled!</h3>
  <p style="margin: 5px 0 0 0;">Colab ‡∏à‡∏∞‡πÑ‡∏°‡πà disconnect ‡∏≠‡∏±‡∏ï‡πÇ‡∏ô‡∏°‡∏±‡∏ï‡∏¥ - ‡∏™‡∏ñ‡∏≤‡∏ô‡∏∞‡∏à‡∏∞‡∏≠‡∏±‡∏û‡πÄ‡∏î‡∏ó‡∏Ç‡πâ‡∏≤‡∏á‡∏ö‡∏ô‡∏ó‡∏∏‡∏Å 5 ‡∏ô‡∏≤‡∏ó‡∏µ</p>
</div>
"""

display(HTML(anti_timeout_script))

print("\n" + "="*70)
print("üõ°Ô∏è  ANTI-TIMEOUT PROTECTION ACTIVATED")
print("="*70)
print("‚úÖ Keep-alive script ‡∏Å‡∏≥‡∏•‡∏±‡∏á‡∏ó‡∏≥‡∏á‡∏≤‡∏ô")
print("‚è∞ ‡∏à‡∏∞‡∏™‡πà‡∏á‡∏™‡∏±‡∏ç‡∏ç‡∏≤‡∏ì‡∏ó‡∏∏‡∏Å 5 ‡∏ô‡∏≤‡∏ó‡∏µ‡∏≠‡∏±‡∏ï‡πÇ‡∏ô‡∏°‡∏±‡∏ï‡∏¥")
print("üí° ‡πÄ‡∏õ‡∏¥‡∏î‡πÅ‡∏ó‡πá‡∏ö‡∏ô‡∏µ‡πâ‡πÑ‡∏ß‡πâ (‡∏¢‡πà‡∏≠ browser ‡πÑ‡∏î‡πâ)")
print("üö´ ‡∏≠‡∏¢‡πà‡∏≤‡∏õ‡∏¥‡∏î‡πÅ‡∏ó‡πá‡∏ö‡∏ô‡∏µ‡πâ ‡∏à‡∏∞‡∏ó‡∏≥‡πÉ‡∏´‡πâ script ‡∏´‡∏¢‡∏∏‡∏î‡∏ó‡∏≥‡∏á‡∏≤‡∏ô")
print("="*70)
```

```python
#@title ## ü§ñ Cell 3: Load Flux Model { display-mode: "form" }

#@markdown ---
#@markdown ### ‡πÇ‡∏´‡∏•‡∏î Flux AI Model
#@markdown ‡πÄ‡∏•‡∏∑‡∏≠‡∏Å Model ‡∏ó‡∏µ‡πà‡∏ï‡πâ‡∏≠‡∏á‡∏Å‡∏≤‡∏£:

MODEL_NAME = "black-forest-labs/FLUX.1-schnell"  #@param ["black-forest-labs/FLUX.1-schnell", "black-forest-labs/FLUX.1-dev"]

#@markdown **‡∏´‡∏°‡∏≤‡∏¢‡πÄ‡∏´‡∏ï‡∏∏:**
#@markdown - `FLUX.1-schnell` = ‡πÄ‡∏£‡πá‡∏ß (4 steps) ‡πÅ‡∏ï‡πà‡∏Ñ‡∏∏‡∏ì‡∏†‡∏≤‡∏û‡∏ï‡πà‡∏≥‡∏Å‡∏ß‡πà‡∏≤‡∏ô‡∏¥‡∏î‡∏´‡∏ô‡πà‡∏≠‡∏¢
#@markdown - `FLUX.1-dev` = ‡∏ä‡πâ‡∏≤‡∏Å‡∏ß‡πà‡∏≤ (20-50 steps) ‡πÅ‡∏ï‡πà‡∏Ñ‡∏∏‡∏ì‡∏†‡∏≤‡∏û‡∏™‡∏π‡∏á‡∏Å‡∏ß‡πà‡∏≤

from diffusers import FluxPipeline
import torch
import gc

print("ü§ñ ‡∏Å‡∏≥‡∏•‡∏±‡∏á‡πÇ‡∏´‡∏•‡∏î Flux Model...")
print(f"üì¶ Model: {MODEL_NAME}")
print("‚è≥ ‡∏≠‡∏≤‡∏à‡πÉ‡∏ä‡πâ‡πÄ‡∏ß‡∏•‡∏≤ 3-5 ‡∏ô‡∏≤‡∏ó‡∏µ (‡πÇ‡∏´‡∏•‡∏î‡∏Ñ‡∏£‡∏±‡πâ‡∏á‡πÄ‡∏î‡∏µ‡∏¢‡∏ß)\n")

# ‡∏•‡πâ‡∏≤‡∏á VRAM ‡∏Å‡πà‡∏≠‡∏ô‡πÇ‡∏´‡∏•‡∏î
gc.collect()
torch.cuda.empty_cache()

# ‡πÇ‡∏´‡∏•‡∏î model
try:
    pipe = FluxPipeline.from_pretrained(
        MODEL_NAME,
        torch_dtype=torch.float16
    )
    pipe = pipe.to("cuda")
    
    # Enable memory efficient attention
    pipe.enable_attention_slicing()
    
    print("\n" + "="*70)
    print("‚úÖ ‡πÇ‡∏´‡∏•‡∏î Flux Model ‡∏™‡∏≥‡πÄ‡∏£‡πá‡∏à!")
    print("="*70)
    print(f"üéØ Model: {MODEL_NAME}")
    print(f"üíæ Device: {pipe.device}")
    print(f"üî¢ Dtype: {pipe.dtype}")
    print("="*70)
    print("üìù ‡∏ï‡πà‡∏≠‡πÑ‡∏õ: ‡∏£‡∏±‡∏ô Cell 4 ‡πÄ‡∏û‡∏∑‡πà‡∏≠‡πÄ‡∏£‡∏¥‡πà‡∏° API Server")
    print("="*70)
    
except Exception as e:
    print(f"\n‚ùå ‡πÄ‡∏Å‡∏¥‡∏î‡∏Ç‡πâ‡∏≠‡∏ú‡∏¥‡∏î‡∏û‡∏•‡∏≤‡∏î: {e}")
    print("\nüí° ‡∏ß‡∏¥‡∏ò‡∏µ‡πÅ‡∏Å‡πâ:")
    print("   1. Runtime > Restart Runtime")
    print("   2. ‡∏£‡∏±‡∏ô Cell ‡πÉ‡∏´‡∏°‡πà‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î")
    print("   3. ‡∏´‡∏£‡∏∑‡∏≠‡∏•‡∏≠‡∏á‡πÄ‡∏õ‡∏•‡∏µ‡πà‡∏¢‡∏ô Model ‡πÄ‡∏õ‡πá‡∏ô schnell (‡πÄ‡∏ö‡∏≤‡∏Å‡∏ß‡πà‡∏≤)")
```

```python
#@title ## üåê Cell 4: Start API Server with Keep-Alive { display-mode: "form" }

#@markdown ---
#@markdown ### ‡πÄ‡∏£‡∏¥‡πà‡∏° API Server ‡∏û‡∏£‡πâ‡∏≠‡∏° Auto Keep-Alive

NGROK_TOKEN = "33jslJ7qBdnf0vJTveRBmj7giBT_5Kp61JVP5iYwmoms2ud9Z"  #@param {type:"string"}
PORT = 5000  #@param {type:"integer"}

#@markdown ---

from flask import Flask, request, jsonify
from flask_cors import CORS
from pyngrok import ngrok
import threading
import base64
from io import BytesIO
from IPython.display import display, HTML
import time
from datetime import datetime

app = Flask(__name__)
CORS(app)

# ‡∏™‡∏ñ‡∏¥‡∏ï‡∏¥‡∏Å‡∏≤‡∏£‡πÉ‡∏ä‡πâ‡∏á‡∏≤‡∏ô
stats = {
    'total_requests': 0,
    'successful_generations': 0,
    'failed_generations': 0,
    'start_time': datetime.now(),
    'last_request': None
}

@app.route('/health', methods=['GET'])
def health():
    """Health check endpoint"""
    uptime = datetime.now() - stats['start_time']
    return jsonify({
        'status': 'healthy',
        'model': MODEL_NAME,
        'uptime_seconds': uptime.total_seconds(),
        'stats': {
            'total_requests': stats['total_requests'],
            'successful': stats['successful_generations'],
            'failed': stats['failed_generations'],
            'last_request': stats['last_request'].isoformat() if stats['last_request'] else None
        }
    })

@app.route('/generate', methods=['POST', 'OPTIONS'])
def generate():
    """Generate image from prompt"""
    if request.method == 'OPTIONS':
        return '', 204
    
    try:
        stats['total_requests'] += 1
        stats['last_request'] = datetime.now()
        
        data = request.json
        prompt = data.get('prompt', '')
        negative_prompt = data.get('negative_prompt', '')
        steps = data.get('steps', 4)  # schnell = 4, dev = 20-50
        width = data.get('width', 512)
        height = data.get('height', 512)
        guidance_scale = data.get('guidance_scale', 0.0)  # schnell ‡πÑ‡∏°‡πà‡πÉ‡∏ä‡πâ guidance
        
        if not prompt:
            return jsonify({'success': False, 'error': '‡∏Å‡∏£‡∏∏‡∏ì‡∏≤‡πÉ‡∏™‡πà prompt'}), 400
        
        print(f"\nüé® ‡∏Å‡∏≥‡∏•‡∏±‡∏á‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏£‡∏π‡∏õ: {prompt[:50]}...")
        print(f"üìê ‡∏Ç‡∏ô‡∏≤‡∏î: {width}x{height} | Steps: {steps}")
        
        # ‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏£‡∏π‡∏õ‡∏†‡∏≤‡∏û
        start_time = time.time()
        
        image = pipe(
            prompt,
            num_inference_steps=steps,
            height=height,
            width=width,
            guidance_scale=guidance_scale
        ).images[0]
        
        generation_time = time.time() - start_time
        
        # ‡πÅ‡∏õ‡∏•‡∏á‡πÄ‡∏õ‡πá‡∏ô base64
        buffered = BytesIO()
        image.save(buffered, format="PNG")
        img_base64 = base64.b64encode(buffered.getvalue()).decode()
        
        stats['successful_generations'] += 1
        
        print(f"‚úÖ ‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏£‡∏π‡∏õ‡∏™‡∏≥‡πÄ‡∏£‡πá‡∏à! ‡πÉ‡∏ä‡πâ‡πÄ‡∏ß‡∏•‡∏≤ {generation_time:.2f} ‡∏ß‡∏¥‡∏ô‡∏≤‡∏ó‡∏µ")
        
        return jsonify({
            'success': True,
            'image': f'data:image/png;base64,{img_base64}',
            'info': {
                'generation_time': generation_time,
                'model': MODEL_NAME,
                'steps': steps,
                'size': f'{width}x{height}'
            },
            'parameters': {
                'prompt': prompt,
                'negative_prompt': negative_prompt,
                'steps': steps,
                'width': width,
                'height': height
            }
        })
        
    except Exception as e:
        stats['failed_generations'] += 1
        print(f"‚ùå Error: {str(e)}")
        return jsonify({
            'success': False,
            'error': str(e)
        }), 500

@app.route('/', methods=['GET'])
def index():
    """API Info"""
    return jsonify({
        'name': 'Flux AI API Server',
        'version': '2.0.0',
        'model': MODEL_NAME,
        'endpoints': {
            'health': 'GET /health',
            'generate': 'POST /generate'
        },
        'status': 'running'
    })

# Keep-alive function
def keep_alive():
    """‡∏™‡πà‡∏á ping ‡∏ó‡∏∏‡∏Å 5 ‡∏ô‡∏≤‡∏ó‡∏µ‡πÄ‡∏û‡∏∑‡πà‡∏≠‡πÑ‡∏°‡πà‡πÉ‡∏´‡πâ ngrok timeout"""
    import requests
    while True:
        try:
            time.sleep(300)  # 5 ‡∏ô‡∏≤‡∏ó‡∏µ
            requests.get(f'http://localhost:{PORT}/health', timeout=5)
            print(f"üíì Keep-alive ping sent at {datetime.now().strftime('%H:%M:%S')}")
        except Exception as e:
            print(f"‚ö†Ô∏è Keep-alive error: {e}")

# ‡πÄ‡∏£‡∏¥‡πà‡∏° ngrok
print("üåê ‡∏Å‡∏≥‡∏•‡∏±‡∏á‡πÄ‡∏£‡∏¥‡πà‡∏° API Server...\n")

try:
    # ‡∏ï‡∏±‡πâ‡∏á‡∏Ñ‡πà‡∏≤ ngrok
    ngrok.set_auth_token(NGROK_TOKEN)
    
    # ‡πÄ‡∏£‡∏¥‡πà‡∏° Flask server ‡πÉ‡∏ô thread
    server_thread = threading.Thread(
        target=lambda: app.run(host='0.0.0.0', port=PORT, debug=False, use_reloader=False)
    )
    server_thread.daemon = True
    server_thread.start()
    
    # ‡∏£‡∏≠‡πÉ‡∏´‡πâ server ‡πÄ‡∏£‡∏¥‡πà‡∏°‡∏ó‡∏≥‡∏á‡∏≤‡∏ô
    time.sleep(3)
    
    # ‡πÄ‡∏£‡∏¥‡πà‡∏° ngrok tunnel
    public_url = ngrok.connect(PORT)
    
    # ‡πÄ‡∏£‡∏¥‡πà‡∏° keep-alive thread
    keepalive_thread = threading.Thread(target=keep_alive)
    keepalive_thread.daemon = True
    keepalive_thread.start()
    
    # ‡πÅ‡∏™‡∏î‡∏á‡∏ú‡∏•‡∏•‡∏±‡∏û‡∏ò‡πå
    display(HTML(f'''
    <div style="background: linear-gradient(135deg, #667eea 0%, #764ba2 100%); padding: 30px; border-radius: 15px; color: white; font-family: Arial; margin: 20px 0;">
        <h1 style="margin: 0 0 20px 0; font-size: 36px;">üéâ API Server ‡∏û‡∏£‡πâ‡∏≠‡∏°‡πÉ‡∏ä‡πâ‡∏á‡∏≤‡∏ô!</h1>
        
        <div style="background: rgba(255,255,255,0.15); padding: 20px; border-radius: 10px; margin: 20px 0;">
            <h2 style="margin: 0 0 15px 0; font-size: 20px;">üåê Public URL</h2>
            <div style="background: rgba(0,0,0,0.3); padding: 15px; border-radius: 8px; font-family: monospace; font-size: 16px; word-break: break-all;">
                {public_url}
            </div>
        </div>
        
        <div style="background: rgba(255,255,255,0.15); padding: 20px; border-radius: 10px; margin: 20px 0;">
            <h3 style="margin: 0 0 15px 0; font-size: 18px;">üõ°Ô∏è ‡∏ü‡∏µ‡πÄ‡∏à‡∏≠‡∏£‡πå‡∏ó‡∏µ‡πà‡πÄ‡∏õ‡∏¥‡∏î‡πÉ‡∏ä‡πâ‡∏á‡∏≤‡∏ô:</h3>
            <ul style="margin: 0; padding-left: 20px; line-height: 1.8;">
                <li>‚úÖ Anti-Timeout Protection (‡∏à‡∏≤‡∏Å Cell 1)</li>
                <li>‚úÖ Auto Keep-Alive Ping (‡∏ó‡∏∏‡∏Å 5 ‡∏ô‡∏≤‡∏ó‡∏µ)</li>
                <li>‚úÖ ngrok Tunnel (Public Access)</li>
                <li>‚úÖ Health Monitoring</li>
                <li>‚úÖ Statistics Tracking</li>
            </ul>
        </div>
        
        <div style="background: rgba(255,255,255,0.15); padding: 20px; border-radius: 10px; margin: 20px 0;">
            <h3 style="margin: 0 0 15px 0;">üìù ‡∏Ç‡∏±‡πâ‡∏ô‡∏ï‡∏≠‡∏ô‡∏ï‡πà‡∏≠‡πÑ‡∏õ:</h3>
            <ol style="margin: 0; padding-left: 20px; line-height: 1.8;">
                <li>‡∏Ñ‡∏±‡∏î‡∏•‡∏≠‡∏Å URL ‡∏î‡πâ‡∏≤‡∏ô‡∏ö‡∏ô</li>
                <li>‡πÄ‡∏õ‡∏¥‡∏î‡πÄ‡∏ß‡πá‡∏ö‡πÑ‡∏ã‡∏ï‡πå‡∏Ç‡∏≠‡∏á‡∏Ñ‡∏∏‡∏ì</li>
                <li>‡πÑ‡∏õ‡∏ó‡∏µ‡πà Settings ‚öôÔ∏è</li>
                <li>‡∏ß‡∏≤‡∏á URL ‡∏•‡∏á‡πÉ‡∏ô "API Endpoint"</li>
                <li>‡∏Å‡∏î Save Settings</li>
                <li>‡∏ó‡∏î‡∏™‡∏≠‡∏ö‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏£‡∏π‡∏õ!</li>
            </ol>
        </div>
        
        <div style="background: rgba(52, 211, 153, 0.2); padding: 15px; border-radius: 8px; border-left: 4px solid #10b981;">
            <p style="margin: 0; font-size: 14px;">
                üí° <strong>Tips:</strong> ‡∏ó‡∏¥‡πâ‡∏á‡πÅ‡∏ó‡πá‡∏ö‡∏ô‡∏µ‡πâ‡πÑ‡∏ß‡πâ (‡∏¢‡πà‡∏≠‡πÑ‡∏î‡πâ) - Anti-Timeout ‡∏à‡∏∞‡∏î‡∏π‡πÅ‡∏•‡πÉ‡∏´‡πâ‡πÑ‡∏°‡πà disconnect
            </p>
        </div>
    </div>
    '''))
    
    print("\n" + "="*70)
    print("üîó API ENDPOINTS:")
    print("="*70)
    print(f"   Health:    {public_url}/health")
    print(f"   Generate:  {public_url}/generate (POST)")
    print("="*70)
    print("\nüíì Keep-alive ping: Active (‡∏ó‡∏∏‡∏Å 5 ‡∏ô‡∏≤‡∏ó‡∏µ)")
    print("üõ°Ô∏è Anti-timeout: Active (‡∏à‡∏≤‡∏Å Cell 1)")
    print("üü¢ Status: Running\n")
    print("="*70)
    print("‚ö†Ô∏è  ‡∏≠‡∏¢‡πà‡∏≤‡∏õ‡∏¥‡∏î‡πÅ‡∏ó‡πá‡∏ö‡∏ô‡∏µ‡πâ! ‡∏à‡∏∞‡∏ó‡∏≥‡πÉ‡∏´‡πâ API Server ‡∏´‡∏¢‡∏∏‡∏î‡∏ó‡∏≥‡∏á‡∏≤‡∏ô")
    print("‚úÖ ‡∏¢‡πà‡∏≠ browser ‡πÑ‡∏î‡πâ ‡πÅ‡∏ï‡πà‡∏≠‡∏¢‡πà‡∏≤‡∏õ‡∏¥‡∏î‡πÅ‡∏ó‡πá‡∏ö")
    print("="*70)
    
except Exception as e:
    print(f"\n‚ùå ‡πÄ‡∏Å‡∏¥‡∏î‡∏Ç‡πâ‡∏≠‡∏ú‡∏¥‡∏î‡∏û‡∏•‡∏≤‡∏î: {e}")
    print("\nüí° ‡∏ß‡∏¥‡∏ò‡∏µ‡πÅ‡∏Å‡πâ:")
    print("   1. ‡∏ï‡∏£‡∏ß‡∏à‡∏™‡∏≠‡∏ö NGROK_TOKEN ‡∏ß‡πà‡∏≤‡∏ñ‡∏π‡∏Å‡∏ï‡πâ‡∏≠‡∏á")
    print("   2. ‡∏•‡∏≠‡∏á‡∏£‡∏±‡∏ô Cell ‡∏ô‡∏µ‡πâ‡∏≠‡∏µ‡∏Å‡∏Ñ‡∏£‡∏±‡πâ‡∏á")
    print("   3. ‡∏´‡∏£‡∏∑‡∏≠ Runtime > Restart Runtime ‡πÅ‡∏•‡πâ‡∏ß‡∏£‡∏±‡∏ô‡πÉ‡∏´‡∏°‡πà‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î")
```
